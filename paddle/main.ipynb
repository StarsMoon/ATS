```python
import os
import paddle
import paddle.nn as nn
import paddlenlp
import numpy as np
# from T5.model import T5ForUnsupervisedGeneration
from T5.model_v2 import T5ForFewshotGeneration
from T5.utils import set_seed
```

```python
class CFG:
    model_name_or_path = "PLM/mengzi-t5/"
    output_dir = "outputs/v2/"
    is_zh = True
    num_saves_kept = 3
    max_source_length = 128
    max_target_length = 32
    num_beams = 4
    train_batch_size = 4
    eval_batch_size = 12
    num_eval_batch = -1#50
    gradient_accumulation_steps = 1
    pretrain_steps = 500
    gold_steps = 5
    learning_rate = 5e-5
    weight_decay = 0.01
    adam_epsilon = 1e-6
    max_grad_norm = 2.0
    num_train_epochs = 3
    max_train_steps = 100000
    warmup_ratio = 0.
    warmup_steps = 0
    lambda_idt = 0.2
    lambda_cyc = 1
    logging_steps = 25
    save_steps = 100
    seed = 2022
    scheduler_type = "cosine"
    use_amp = True
    scale_loss = 2 ** 15
    logdir = "logs"
    num_workers = 4
    ckp_path = "outputs/v2/best"
    init_step = 0

CFG.logdir = os.path.join(CFG.output_dir, CFG.logdir)
if not os.path.exists(CFG.model_name_or_path):
    CFG.model_name_or_path = "Langboat/mengzi-t5-base"
set_seed(CFG.seed)
```

```python
paddle.set_device('gpu')
model = T5ForFewshotGeneration(CFG)
```

```python
model.load(CFG.ckp_path)
decoded_preds, decoded_labels = model.predict()
with open("outputs/v2/best/predictions.txt", "w", encoding="utf-8") as f:
    for pred, label in zip(decoded_preds, decoded_labels):
        f.write(pred + "\n" + label + "\n")
```

```python
model.finetune(pretrain=True)
```
